## 3. Modelling

Ideally I would try to train a model per product or a cluster of products. For the latest, I would train a SVM model  for each of the products and then cluster them through the model parameters. Due to the time and computing limitation I will skip this step and train just one model with all the data available for all the products.

The code will performe RandomizeSearchCV for a list of models and parameters to choose the optimal ones. 

The metric chosen is the Root Mean Squared Log Error, this score provides robustness to treat outliers, it is relative error making that this no scale when the there is a high difference. 

Unfortunatelly, due to a limitation in the computing power I was not able to tunne and train the models, so insted I perfomed a basid Linear Regression. For future steps this one will serve as a baseline in model competition.
